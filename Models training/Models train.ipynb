```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/3w/all_features.csv')
```

```python
data
```

```python
# import pycaret classification and init setup
from pycaret.classification import *
setup(data, target = 'class', session_id = 123)
```

# teste 01


```python
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
type(exp)
```

```python
# init setup on exp
exp.setup(data, target = 'class', session_id = 123)
```

```python
# compare baseline models
best = compare_models(fold=5, include = ['rf', 'et', 'lightgbm'])
```

```python
# plot confusion matrix
plot_model(best, plot = 'confusion_matrix')
```

```python
# plot AUC
plot_model(best, plot = 'auc')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
evaluate_model(best)
```

```python
rf = create_model('rf', fold=5)
```

```python
plot_model(rf, plot = 'confusion_matrix', plot_kwargs = {'percent' : True})
```
